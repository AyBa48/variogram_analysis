# Continental Terminal 3 Data

Read and display summary statistic

```python
###Read ct3 data
df=pd.read_excel('CT3_Kriging_Data.xlsx')
```

#data[['column1', 'column2', 'column3']].describe()

```python
#data statistic
df[['NS_m_calc','SRTM_v3_30','Piezo_AB', 'Distance']].describe()
```

```python
# visualization of the PWL vs Year of measurement
plt.figure(figsize=(14,7))
plt.scatter(df['Piezo_AB'],df['NS_annee'],   label='Borehole')
#plt.plot(df['SRTM_v3_30'], label='Surface Elevation')
#plt.plot(df['NS_m_calc'], label='Groundwater Table Depth')
plt.legend(loc='upper left')
plt.xlabel('Piezometric Water Level')
plt.ylabel('Year of measurement')
plt.grid(True)
```

   - GroundWater level is estimated by soustractring Aquifer Static level from the Elevation of the Soil Surface
   - The data show that when estimating the water level, the resulting values follow a trend in the direction of the Elevation
   - in the extrem West of the region of interest the Elevation of the Soil surface start to diverge sligthly from the global mean. The abrupt changes in the Elevation data is transposed into to the water level. 

```python
#data before 1970 -- between 1970 - 1990 and --- from 1990 to 2019
d70=df[df['NS_annee']<=1970]
d70=d70.reset_index(drop=True)
d70.to_excel('CT3_data_bef70.xlsx')
d90=df[(df['NS_annee']<= 1990) & (df['NS_annee']>1970)]
d90=d90.reset_index(drop=True)
d90.to_excel('CT3_data_bef90.xlsx')
d19=df[df['NS_annee']>1990]
d19=d19.reset_index(drop=True)
d19.to_excel('CT3_data_bef19.xlsx')

#Plot 
#plt.figure(figsize=(14,7))
plt.subplot(221)
plt.plot(d70['Piezo_AB'],   label='Variation of the Piezometric water level')
#plt.plot(df['SRTM_v3_30'], label='Surface Elevation')
#plt.plot(df['NS_m_calc'], label='Groundwater Table Depth')
plt.legend(loc='upper left')
plt.xlabel('Number of Borehole')
plt.ylabel('Piezometric Water Level')
plt.title('Piezometric Data before 1970')

plt.grid(True)
plt.subplot(222)
plt.plot(d90['Piezo_AB'],   label='Variation of the Piezometric water level')
#plt.plot(df['SRTM_v3_30'], label='Surface Elevation')
#plt.plot(df['NS_m_calc'], label='Groundwater Table Depth')
plt.legend(loc='upper left')
plt.xlabel('Number of Borehole')
plt.ylabel('Piezometric Water Level')
plt.title('Piezometric Data between 1970 and 1990')
plt.grid(True)
plt.subplot(223)
plt.plot(d19['Piezo_AB'],   label='Variation of the Piezometric water level')
#plt.plot(df['SRTM_v3_30'], label='Surface Elevation')
#plt.plot(df['NS_m_calc'], label='Groundwater Table Depth')
plt.legend(loc='upper left')
plt.xlabel('Number of Borehole')
plt.ylabel('Piezometric Water Level')
plt.title('Piezometric Data between 1990 and 2019')
plt.grid(True)
plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=2.8, wspace=0.2, hspace=0.3)
plt.show()
```
